```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image1 = cv2.imread("chien.jpg")
image2 = cv2.imread("chien.jpg")
```

```python
# Convert to RGB
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

```python
# Reshape the image to a 2D array of pixels and 3 color values (RGB)
pixel_values1 = image1.reshape((-1,3))
pixel_values2 = image2.reshape((-1,3))
# Convert to float
pixel_values1 = np.float32(pixel_values1)
pixel_values2 = np.float32(pixel_values2)
print(pixel_values1.shape)
print(pixel_values2.shape)
```

```python
# Define stopping criteria
criteria1 = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
criteria2 = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
```

```python
# Number of clusters (k)
K1 = 3
_1, labels1, (centers1) = cv2.kmeans(pixel_values1, K1, None, criteria1, 10, cv2.KMEANS_RANDOM_CENTERS)
K2 = 2
_2, labels2, (centers2) = cv2.kmeans(pixel_values2, K2, None, criteria2, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
K = [2,3,4]
```

```python
# Convert back to a 8 bit values
centers1 = np.uint8(centers1)
centers2 = np.uint8(centers2)
# Flatten the labels array
labels1 = labels1.flatten()
labels2 = labels2.flatten()
```

```python
# Convert all pixels to the color of the centroids
segmented_image1 =  centers1[labels1.flatten()]
segmented_image2 =  centers2[labels2.flatten()]
```

```python
# Reshape back to the original image dimension
segmented_image1 = segmented_image1.reshape(image1.shape)
# show the image
plt.imshow(image1)
plt.show()
plt.imshow(segmented_image1)
plt.show()
```

```python
# Reshape back to the original image dimension
segmented_image2 = segmented_image2.reshape(image2.shape)
# show the image
plt.imshow(image2)
plt.show()
plt.imshow(segmented_image2)
plt.show()
```
