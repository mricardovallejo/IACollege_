## 3) Transformation de donnes

```python
df2 = df.copy()
df2['SEX'] = df2['SEX'].replace([0, 1], ['M', 'F'])
df2['SMOKER'] = df2['SMOKER'].replace([0, 1], ['Y', 'N'])
df2['TIME'] = df2['TIME'].replace([0, 1], ['J', 'S'])
df2['DAY'] = df2['DAY'].replace([1, 2, 3, 4, 5, 6, 7], ['Lu', 'Ma', 'Mi', 'Ju', 'Vi', 'Sa', 'Di'])

 # Pour eviter le bias sur les gros commandes et reduire l information des multiples variables.
df2['porc_tip_unit'] = ( (df2['TIP'] * 100) /  df2['TOTBILL'] ) /  df2['SIZE'] 
```

```python

df2.drop('TOTBILL', axis = 'columns', inplace=True)
df2.drop('SIZE', axis = 'columns', inplace=True)
df2.drop('TIP', axis = 'columns', inplace=True)
df2
```

```python
df2['porc_tip_unit'].describe()
```

```python
bins = [0,10, 20, 30]  # Pourcentage unitaire de pourboire
labels = ['BAS', ' MEDIUM', 'GOOD']
```

```python
df2
```



Pour example TOTAL BILL 75
Pourboire 5
Numero unites 3

( 5 * 100 / 75 ) / N --> Pourboir pourcentual unitaire.

```python
df2['porc_tip_unit'] = pd.cut(df2['porc_tip_unit'], bins, labels=labels)
```

```python
df2.tail(50)
```

```python
df2.set_index("IDEN", inplace = True)
```

```python
df3  = pd.get_dummies(df2)
df3
```

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt


freq_items = apriori(df3, min_support=0.2, use_colnames=True, verbose=1)
freq_items
```

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)
rules.head(50)
```

```python
rules.tail(50)
```

# Conclusions

Pour detecter la predisposition que donne plus de pourboirs on analyse les antecedents ou le consequent est:
porc_tip_unit_GOOD ou porc_tip_unit_BAS, de tal facon que on detect les conditions que produisent les bonne pourboirs.



```python
rules.sort_values(by="confidence", ascending=False ).head(30)
```

En observation des regles nous dit:
Samedis et vendredis, le sex masculine, ne donne pas de bonne pourboire avec 87.9% de confiance.
Les smokers, les samedis et les soires ne sont pas bonne pour les pourboires.

On a plein de commandes sans bonne pourboire, donc on a pas regles d association pour le bonne pourboir mais on peut identifier
les conditions pour pas bonne pourboire.

On a pas suffisant de donnes avec bonne pourboire pour caracteriser, mais on a une bonne caracterisation de mauvaise pourboire.
