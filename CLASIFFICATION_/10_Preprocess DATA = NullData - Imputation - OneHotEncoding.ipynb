##  2.7  Detection de valeurs aberrantes - Z score

```python

# random data points to calculate z-score
data = tit2['prix']
# calculate mean
mean = np.mean(data)
print ("mean: ", mean)
# calculate standard deviation
sd = np.std(data)

threshold = 2

outliers = []

for i in data: 
    z = (i-mean)/sd # calculate z-score
    if abs(z) > threshold:  # identify outliers
        outliers.append(i) # add to the empty list
 
print("The detected outliers are: ", outliers) 
```

```python
threshold = 3

outliers = []

for i in data: 
    z = (i-mean)/sd # calculate z-score
    if abs(z) > threshold:  # identify outliers
        outliers.append(i) # add to the empty list
  
print("The detected outliers are: ", outliers) 
```

# 5. Encodage - OneHot

```python
from sklearn.preprocessing import OneHotEncoder

tit6 = titnotNullQualitative.copy()  #With preprocessed data
display(tit6)
oneHot_encoder = OneHotEncoder(sparse=False)

label_encoder = LabelEncoder()
labelencoded = label_encoder.fit_transform(tit6['abb'])
labelencoded = labelencoded.reshape(len(tit6), 1)

tit_oneHotEncoded= oneHot_encoder.fit_transform(labelencoded)
tit_oneHotEncodedDF = pd.DataFrame(data=tit_oneHotEncoded) #To convert to Dataframe
display("abb encoded:", tit_oneHotEncodedDF)

```

```python
# Integrate Coding to DATAFRAME to Process

#Create headers columns
titCountyrMap = tit6[['abb']].copy().dropna().drop_duplicates().reset_index(drop=True)
encodeheaders = titCountyrMap.sort_values('abb', ascending=True)  #Its sorting mandatory because onehot algoritm
print(encodeheaders)

tit_oneHotEncodedDF.columns = encodeheaders
display(tit_oneHotEncodedDF)
```

```python
#Concatenate dataFrames, axis=1 cest pour concat horizontal
result = pd.concat([tit6, tit_oneHotEncodedDF], axis=1, sort=False)
display(result)
```
