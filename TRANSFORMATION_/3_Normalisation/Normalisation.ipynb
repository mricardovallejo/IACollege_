# Excercise 1

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

np.random.seed(1)
#Création de dataframe avec 3 series distribution Loi Normale
df = pd.DataFrame({'col1':np.random.normal(0,2,10000) ,
                  'col2': np.random.normal(5,3,10000),    #mu=5 sigma=3  mu cest le centre de la distribution nsamples=1000
                  'col3': np.random.normal(-5,5,10000)
                  })

```

```python
#Creation du Standard Scaler - Data Standardization
scaler = preprocessing.StandardScaler()
df_echelle =scaler.fit_transform(df)
#conversion en Dataframe
df_echelle = pd.DataFrame(df_echelle, columns=['col1','col2','col3'])
```

```python
#Preparation des plots
figure, (cx1, cx2) = plt.subplots(ncols=2, figsize=(6,5))

cx1.set_title("Avant mise à l'échelle")
sns.kdeplot(df['col1'], ax=cx1)
sns.kdeplot(df['col2'], ax=cx1)
sns.kdeplot(df['col3'], ax=cx1)

cx2.set_title("Apres mise à l'échelle")
sns.kdeplot(df_echelle['col1'], ax=cx2)
sns.kdeplot(df_echelle['col2'], ax=cx2)
sns.kdeplot(df_echelle['col3'], ax=cx2)
plt.show()
```

# Excercise 2

```python
#Création de dataframe avec 3 series distribution Loi Normale
df = pd.DataFrame({'col1':np.concatenate([np.random.normal(20,1,10000),
                                         np.random.normal(1,1,500)]) ,
                  'col2': np.concatenate([np.random.normal(30,1,10000),
                                         np.random.normal(50,1,500)]) 
                  })
```

```python
#Creation du robustScaler
scaler = preprocessing.RobustScaler()
#scaler = preprocessing.StandardScaler()
#scaler = preprocessing.MinMaxScaler()
df_echelle =scaler.fit_transform(df)
#conversion en Dataframe
df_echelle = pd.DataFrame(df_echelle, columns=['col1','col2'])
```

```python
#Preparation des plots
figure, (cx1, cx2) = plt.subplots(ncols=2, figsize=(6,5))

cx1.set_title("Avant mise à l'échelle")
sns.kdeplot(df['col1'], ax=cx1)
sns.kdeplot(df['col2'], ax=cx1)

cx2.set_title("Apres mise à l'échelle")
sns.kdeplot(df_echelle['col1'], ax=cx2)
sns.kdeplot(df_echelle['col2'], ax=cx2)

plt.show()
```

# Excercise 3 -  Distributions differentes

```python
#Création de dataframe avec 3 series distribution Loi Normale
df = pd.DataFrame({'col1':np.random.chisquare(8,1000),
                  'col2': np.random.beta(8,2,1000) * 4, 
                  'col3': np.random.normal(50,3,1000)
                  })
```

```python
#Creation du robustScaler
#scaler = preprocessing.RobustScaler()
scaler = preprocessing.StandardScaler()
#scaler = preprocessing.MinMaxScaler()
df_echelle =scaler.fit_transform(df)
#conversion en Dataframe
df_echelle = pd.DataFrame(df_echelle, columns=['col1','col2', 'col3'])
```

```python
#Preparation des plots
figure, (cx1, cx2) = plt.subplots(ncols=2, figsize=(6,5))

cx1.set_title("Avant mise à l'échelle")
sns.kdeplot(df['col1'], ax=cx1)
sns.kdeplot(df['col2'], ax=cx1)
sns.kdeplot(df['col3'], ax=cx1)

cx2.set_title("Apres mise à l'échelle")
sns.kdeplot(df_echelle['col1'], ax=cx2)
sns.kdeplot(df_echelle['col2'], ax=cx2)
sns.kdeplot(df_echelle['col3'], ax=cx2)

plt.show()
```
