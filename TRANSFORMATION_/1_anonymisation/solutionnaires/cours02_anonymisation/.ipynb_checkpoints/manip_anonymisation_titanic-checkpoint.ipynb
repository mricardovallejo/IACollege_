```python
import pandas as pd
df = pd.read_csv("titanic.csv", sep=";")
df.head()
```

```python
#determiner le type de chaque serie
df.dtypes
```

```python
df.describe()
```

```python
# ["Name"] contient des informations personnelles
#On constate que c'est unique pour chaque passager/ on peut supprimer
len(df.name.unique() )
```

```python
#["Ticket", "Cabin"]contiennent des informations personnelles
#On constate que c'est presque unique pour chaque passager
#Masker 2 caracteres apres les 2 premiers digits dans cabin
df.cabin = df.cabin.map(lambda x: str(x)[0:2] +"***" )
df
```

```python
#Sexe pour etre encodé (pseudonymisé)
#Déclarer la table de correspondance de sexe
d = {'categories':[1,2]}
table_cor = pd.DataFrame(d)
table_cor
```

```python
#anonymiser en effectuant une correspondance
df.sex = df.sex.map(lambda x: table_cor.categories[0] if x == "male" else table_cor.categories[1] 
                    if x == "female" else "")
df
```

```python
#Il y'a des lignes qui contiennent des valeurs manquantes
features=["pclass","survived","name","sex","age","sibsp","parch","ticket","fare","cabin","embarked",
          "boat","body","home.dest"]

for tmp in features: 
    print("Serie:{}, Null:{}".format(tmp,df[tmp].isnull().sum()))
```

```python
#Supprimer les series body  
cols_supp = ["name", "body"]
df.drop(columns=cols_supp, inplace=True)
df.head()
```
