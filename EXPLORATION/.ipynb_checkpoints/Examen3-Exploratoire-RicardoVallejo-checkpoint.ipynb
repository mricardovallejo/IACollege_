# 2

```python
TC = pd.crosstab(dataOriginal.ID,dataOriginal.Product)
TC.head(5) # 5 only for easy print
```

```python
TCx = pd.DataFrame(TC.sum(), columns = ['App_freq'])
Rpta = TCx.sort_values(by='App_freq', ascending=False).head(47)
Rpta
```

```python
plt.rcParams["figure.figsize"] = (20,10)
Rpta.plot(kind="bar")
```
