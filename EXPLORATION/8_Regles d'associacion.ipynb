### 5.En fixant une valeur minimale de support de 0.025, et un cardinal inférieur ou égal à 4, déterminer l’ensemble des items fréquents.

```python
# pip install mlxtend  

#importation de la fonction apriori
from mlxtend.frequent_patterns import apriori
#itemsets frequents
freq_itemsets = apriori(TC,min_support=0.025,max_len=4,use_colnames=True)
freq_itemsets
```

# 7. En fixant un seuil de confiance minimal à 0.75, générer les règles d’association.

Nous produisons les règles à partir des itemsets fréquents. Elles peuvent être très nombreuses,
nous en limitons la prolifération en définissant un seuil minimal (min_threshold = 0.75) sur une
mesure d’intérêt, en l’occurrence la confiance dans notre exemple (metric = ‘’confidence’’).

```python
#fonction de calcul des règles
from mlxtend.frequent_patterns import association_rules
#génération des règles à partir des itemsets fréquents
regles = association_rules(freq_itemsets,metric="confidence",min_threshold=0.75)
regles.head(10)
```

```python
regles.shape # : l’antécédent, le conséquent, et 7 indicateurs numériques d’évaluation des règles
```

```python
regles.columns
```

```python
myRegles = regles[['antecedents','consequents','lift']]
pandas.set_option('precision',3)
```

```python
myRegles.head(5)
```

# Afficher les 10 meilleures règles en se basant sur le lift.

```python
myRegles.sort_values(by='lift', ascending=False).head(10)
myRegles
```

```python
#affichage des règles avec un LIFT supérieur ou égal à 7
print(myRegles[myRegles['lift'].ge(7.0)])
```
